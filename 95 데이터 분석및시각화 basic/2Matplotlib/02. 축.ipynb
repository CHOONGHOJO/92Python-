```python
import matplotlib.pyplot as plt
import matplotlib
# matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트사용시 마이너스 글자 깨지는 현상 해결
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x, y)
plt.title('LINE GRAPH', fontdict={'family':'HyGungSo-Bold','size':20}) # 개별 폰트 설정
```

```python
plt.plot(x, y)
plt.xlabel('X side')
plt.ylabel('Y side')
```

```python
plt.plot(x, y)
plt.xlabel('X side', color='red')
plt.ylabel('Y side', color='#00aa00')
```

```python
plt.plot(x, y)
plt.xlabel('X side', color='red', loc='right') # left, center, right
plt.ylabel('Y side', color='#00aa00', loc='top') # top, center, bottom
```

```python
plt.plot(x, y)
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])
plt.show()
```

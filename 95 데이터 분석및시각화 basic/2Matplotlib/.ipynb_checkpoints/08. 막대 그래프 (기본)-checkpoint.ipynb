# 8. 막대 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트사용시 마이너스 글자 깨지는 현상 해결
```

```python
labels = ['KANG', 'SEO', 'JEON']
values = [190, 187, 184]

plt.bar(labels, values)
```

```python
labels = ['KANG', 'SEO', 'JEON']
values = [190, 187, 184]

plt.bar(labels, values, color='r')
```

```python
labels = ['KANG', 'SEO', 'JEON']
values = [190, 187, 184]
colors = ['r', 'g', 'b']
plt.bar(labels, values, color=colors, alpha=0.5)
```

```python
labels = ['KANG', 'SEO', 'JEON']
values = [190, 187, 184]

plt.bar(labels, values)
plt.ylim(175, 195)
```

```python
plt.bar(labels, values, width=0.5)
```

```python
plt.bar(labels, values, width=0.3)
plt.ylim(175, 195)
plt.xticks(rotation=45) # x축의 label 각도를 45도로 설정
plt.yticks(rotation=45) # y축의 value 각도를 45도로 설정
```

```python
labels = ['KANG', 'SEO', 'JEON']
values = [190, 187, 184]
ticks = ['No.1', 'No.2', 'No.3']
plt.ylim(175, 195)

plt.bar(labels, values)
plt.xticks(labels, ticks, rotation=70)
```
